# Python Data Types

```python
'''
Python has various data types. Here are some of the basic ones:

- Numbers: These can be integers (e.g., 5), floats (e.g., 5.0), or complex numbers.
- Strings: These are sequences of character data, like "hello".
- Lists: These are ordered collections of items (which can be of different types), like [1, "a", 2.0].
- Dictionaries: These are unordered collections of key-value pairs, like {"name": "John", "age": 30}.

Let's try them out.
'''
# Numbers
x = 5 # This is an integer
y = 5.0 # This is a float
print(x, y)

# Strings
s = "hello"
print(s)

# Lists
list_example = [1, "a", 2.0]
print(list_example)

# Dictionaries
dict_example = {"name": "John", "age": 30}
print(dict_example)

```

# Control Flow (if-elif-else)

```python
# Demo: Python provides if, elif, and else statements to control the flow of your programs.
# Let's create a simple grading system as a demonstration.
score = 85 # you can change this value to test different conditions

if score >= 90:
    print('Grade is A')
elif score >= 80:
    print('Grade is B')
else:
    print('Grade is C')

# Exercise 1: Add two more conditions to the grading system to handle 'D' (score >= 60) and 'F' (score < 60) grades.
# You can copy the above code and modify it below.

# Exercise 2: Create a function that accepts a score and returns the grade as a string. Test your function with some values.
```

# Loops (for, while)

```python
# Demo: Python provides for and while loops to iterate over a sequence or repeat a block of code.
# Here is a demo of using a for loop to iterate over a list.

# List of student names
students = ["John", "Mary", "Paul", "Mike", "Emma"]

for student in students:
    print(f"Welcome, {student}!")

# Exercise 1: Modify the loop above to print "Welcome, [student name]!" only if the student's name has more than 3 characters.

# Exercise 2: Write a while loop that prints the squares of numbers from 1 to 5. Be careful with infinite loops!
```

# Errors and Exception Handling

```python
# Demo: Python uses try and except blocks for error and exception handling.
# Here is a simple demo of catching a ZeroDivisionError.

try:
    print(5/0)
except ZeroDivisionError:
    print("You can't divide by zero!")

# Exercise 1: Write a function that accepts two numbers and performs division, handling any ZeroDivisionError exceptions that may occur.

# Exercise 2: Modify the function from Exercise 1 to also handle TypeError exceptions (e.g., if the function is called with non-numeric arguments).
```

# Data Structures: Lists

```python
# Demo: Python provides several data structures, such as lists and dictionaries.
# Here is a demo of some common operations on lists.

numbers = [5, 3, 9, 1] # A list of numbers

# Append a number
numbers.append(7)
print(numbers)

# Remove a number
numbers.remove(3)
print(numbers)

# Sort the list
numbers.sort()
print(numbers)

# Exercise 1: Write a function that accepts a list of numbers and returns a new list with all numbers squared.

# Exercise 2: Write a function that accepts a list of strings and returns a new list that contains only the strings that start with the letter 'a'.
```
